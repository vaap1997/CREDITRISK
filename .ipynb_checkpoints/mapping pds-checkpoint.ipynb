```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('small_companies_chosen_variables.csv')
data = data.iloc[:,1:]
```

```python
data.head()
```

```python
df = pd.read_excel('Rahull_Values.xlsx', sheet_name='Historical_1yrPD_2016_Values')
df.iloc[:,0] = df.iloc[:,0].map(lambda x: x.split()[0])
df.rename(columns={
    'Bloomberg 1 year PD 2023': 'pd23',
    'Bloomberg 1 year PD 2022': 'pd22',
    'Bloomberg 1 year PD 2021': 'pd21',
    'Bloomberg 1 year PD 2020': 'pd20',
    'Bloomberg 1 year PD 2019': 'pd19',
    'Bloomberg 1 year PD 2018': 'pd18',
    'Bloomberg 1 year PD 2017': 'pd17',
    'Bloomberg 1 year PD 2016': 'pd16'}, inplace=True)
```

```python
df.head()
```

```python
data.shape, df.shape
```

```python
data[data['Year']==2014]
```

```python
# common companies
pd.merge(data, df, how='inner', left_on='Ticker', right_on='Company')[['Ticker']].drop_duplicates()
```

```python
merged_data = pd.merge(data, df, how='left', left_on='Ticker', right_on='Company')
merged_data.drop('Company', axis=1, inplace=True)
merged_data.drop_duplicates(inplace=True)
```

```python
merged_data.shape
```

## creating pd_score column with corresponding pd scores for each year

```python
import numpy as np

def score(df):
    if df['Year'] in [2014, 2015]:
        return np.nan
    elif df['Year'] == 2016:
        return df['pd16']
    elif df['Year'] == 2017:
        return df['pd17']
    elif df['Year'] == 2018:
        return df['pd18']
    elif df['Year'] == 2019:
        return df['pd19']
    elif df['Year'] == 2020:
        return df['pd20']
    elif df['Year'] == 2021:
        return df['pd21']
    elif df['Year'] == 2022:
        return df['pd22']
    elif df['Year'] == 2023:
        return df['pd23']

# Create 'pd_score' column using the score function
merged_data['pd_score'] = merged_data.apply(score, axis=1)
```

```python
merged_data.filter(regex='pd\d+').columns
```

```python
merged_data = merged_data.drop(merged_data.filter(regex='pd\d+').columns, axis=1)
```
